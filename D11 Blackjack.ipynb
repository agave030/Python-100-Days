```python
import random
import art
from replit import clear

# 隨機發卡，Ace預設為11
def deal_card():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    return random.choice(cards)

# 計算總和
def calculate_score(list_of_cards):
    score = sum(list_of_cards)
    # 若分數大於21且有Ace，則把Ace的十一點換成一點，再計算總和
    if score > 21 and 11 in list_of_cards:
        list_of_cards.remove(11)
        list_of_cards.append(1)
        return sum(list_of_cards)
    return score

# 加卡並印出卡片項目和總額
def adding_card(user_cards, computer_cards):
    still_continue = True
    # asking for adding other cards
    while still_continue:
        yes_or_no = input("Type 'y' to get another card, type 'n' to pass: ").lower()
        # 在還沒超過21點的情況下不停的加卡片，大於等於21則結束迴圈
        if yes_or_no == "y":
          if calculate_score(user_cards) < 21:
              user_cards.append(deal_card())
          else:
              still_continue = False
          user_scores = calculate_score(user_cards)
          # 印出你現在的卡片與總額
          print(f"Your cards: {user_cards}, current score: {user_scores}")
        elif yes_or_no == 'n':
          print(f"Your cards: {user_cards}, current score: {sum(user_cards)}") 
          still_continue = False
        else:
          print("What are you typing?")
    print(f"Computer cards: {computer_cards}, computer score: {sum(computer_cards)}")

def compare(user_scores, computer_scores):
    if user_scores > 21 or computer_scores == 21:
        return False
    elif computer_scores > 21 or user_scores == 21:
        return True
    elif computer_scores == user_scores:
        return "Draw"
    elif user_scores > computer_scores:
        return True
    else:
        return False

while 1:
      user_cards = []
      computer_cards = []
      user_cards.append(deal_card())
      user_cards.append(deal_card())
      computer_cards.append(deal_card())
      computer_cards.append(deal_card())
      if sum(user_cards) == 21:
        print("You win!")
        continue
      elif sum(computer_cards) == 21:
        print("You lose!")
        continue
      
      print(art.logo)
      choice = input("Do you want to play a game of Blackjack? Type 'y' or 'n': ")
      while calculate_score(computer_cards) < 17:
          computer_cards.append(deal_card())
      if choice == "y":
          print(f"Your cards: {user_cards}, current score: {calculate_score(user_cards)}")
          print(f"Computer first card: {computer_cards[0]}")
          adding_card(user_cards, computer_cards)
          win_or_lose = compare(calculate_score(user_cards), calculate_score(computer_cards))
          if win_or_lose == True:   
              print("You win!")
          elif win_or_lose == False:  
              print("You lose!")
          elif win_or_lose == "Draw":
              print("Draw")
      else:
          break
```
